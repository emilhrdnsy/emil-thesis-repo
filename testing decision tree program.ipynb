```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Memuat dataset Iris
iris = load_iris()
X, y = iris.data, iris.target

# Membuat model Random Forest dengan parameter n_estimators
model = RandomForestClassifier(n_estimators=5, max_depth=9, random_state=42)
model.fit(X, y)

# Visualisasi semua pohon keputusan dari Random Forest
for i, estimator in enumerate(model.estimators_):
    plt.figure(figsize=(20, 18))
    plot_tree(
        estimator,  # Mengakses setiap pohon dalam ensemble
        feature_names=iris.feature_names,
        class_names=iris.target_names,
        filled=True,
        rounded=True
    )
    plt.title(f"Visualisasi Pohon Ke-{i+1} dalam Random Forest")
    plt.show()
```
