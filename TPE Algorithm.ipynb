```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Data for good and bad hyperparameters
good_hyperparameters = np.array([0.01, 0.02])
bad_hyperparameters = np.array([0.04, 0.04, 0.05, 0.06, 0.07, 0.08, 0.09])

# Perform Kernel Density Estimation
kde_good = gaussian_kde(good_hyperparameters, bw_method='scott')
kde_bad = gaussian_kde(bad_hyperparameters, bw_method='scott')

# Generate x values for plotting
x_values = np.linspace(0, 0.1, 1000)

# Compute KDE values for the x values
kde_values_good = kde_good(x_values)
kde_values_bad = kde_bad(x_values)

# Plotting the KDEs
plt.figure(figsize=(10, 6))
plt.plot(x_values, kde_values_good, label='KDE for $D^{(l)}$ (Good Hyperparameters)', color='blue')
plt.plot(x_values, kde_values_bad, label='KDE for $D^{(g)}$ (Bad Hyperparameters)', color='red')
plt.fill_between(x_values, kde_values_good, alpha=0.3, color='blue')
plt.fill_between(x_values, kde_values_bad, alpha=0.3, color='red')
plt.title('Kernel Density Estimation for Hyperparameters')
plt.xlabel('Hyperparameter Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Data observasi
data = np.array([0.01, 0.02, 0.04, 0.04, 0.05, 0.06, 0.07, 0.08, 0.09])

# Estimasi kepadatan dengan KDE
kde = gaussian_kde(data, bw_method='scott')

# Menghasilkan nilai x untuk plot
x_values = np.linspace(0, 0.1, 1000)

# Menghitung nilai KDE untuk nilai x
kde_values = kde(x_values)

# Plot KDE
plt.figure(figsize=(8, 6))
plt.plot(x_values, kde_values, label='KDE', color='blue')
plt.fill_between(x_values, kde_values, alpha=0.5, color='blue')
plt.title('Kernel Density Estimation')
plt.xlabel('Hyperparameter Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```
