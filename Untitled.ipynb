```python
import seaborn as sns
import matplotlib.pyplot as plt

# Memuat dataset
iris = sns.load_dataset('iris')
```

```python
# Histogram dengan jumlah bins default
sns.histplot(data=iris, x='sepal_length')
plt.show()
```

```python
# Histogram dengan 5 bins
sns.histplot(data=iris, x='sepal_length', bins=5)
plt.show()

# Histogram dengan 20 bins
sns.histplot(data=iris, x='sepal_length', bins=20)
plt.show()
```
